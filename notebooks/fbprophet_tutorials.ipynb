# Quick Start

```python
from fbprophet import Prophet
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/master/examples/example_wp_log_peyton_manning.csv')
```

```python
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
# 一年分のデータが追加される
future
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
# yhat       : ???
# yhat_lower : ???
# yhat_upper : ???
```

```python
fig = m.plot(forecast)
```

```python
fig = m.plot_components(forecast)
```

# Saturating Forecast

```python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/master/examples/example_wp_log_R.csv')
df
```

```python
df['cap'] = 8.5
df
```

```python
m = Prophet(growth='logistic')
m.fit(df)
```

```python
future = m.make_future_dataframe(1826)
future['cap'] = 8.5
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
future = m.make_future_dataframe(1826)
# future['cap'] = 8.5
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
future = m.make_future_dataframe(1826)
future['cap'] = 7.5
# futureで指定した値が優先される
forecast = m.predict(future)
fig = m.plot(forecast)
```

```python
df['y'] = 10 - df['y']
df['cap'] = 6
df['floor'] = 1.5
future['cap'] = 6
future['floor'] = 1.5
m = Prophet(growth='logistic')
m.fit(df)
forecast = m.predict(future)
fig = m.plot(forecast)
```

# Trend Changepoint

```python
from fbprophet.plot import add_changepoints_to_plot
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/master/examples/example_wp_log_peyton_manning.csv')
m = Prophet()
m.fit(df)

future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)

fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(), m, forecast)
```
